```python
from Bio import Phylo
```

```python
tree_file = "./m1_mig1e-2_k1e3_f1_p0.3.nwk"
tree = Phylo.read(tree_file, 'newick', rooted=True)
```

```python
with open("m1_mig1e-2_k1e3_f1_p0.3_METlineage.txt") as f1:
    met_lineage = [n.strip() for n in f1.readlines()]
met_lineage
```

```python
regimes = ['node,node2,regime']
for node in tree.get_terminals():
    if node.name in met_lineage:
        regimes.append(node.name + ",,2")
    else:
        regimes.append(node.name + ",,1")
regimes
```

```python
def get_leaf(node):
    if node.is_terminal():
        return node.name
    else:
        return get_leaf(node.clades[0])
```

```python
for node in tree.find_clades():
    if node.is_terminal() or len(node.clades) < 2:
        continue
    leaf1 = get_leaf(node.clades[0])
    leaf2 = get_leaf(node.clades[1])
    if leaf1 in met_lineage and leaf2 in met_lineage:
        regimes.append(','.join([leaf1, leaf2, '2']))
    else:
        regimes.append(','.join([leaf1, leaf2, '1']))
regimes
```

```python
len(regimes)
```

```python
with open("m1_mig1e-2_k1e3_f1_p0.3_METregime.csv", 'w') as f:
    for n in regimes:
        f.write(n + "\n")
```

```python
with open("m1_mig1e-2_k1e3_f1_p0.3_PRIregime.csv", 'w') as f:
    f.write(regimes[0] + "\n")
    for n in regimes[1:]:
        f.write(','.join(n.split(',')[:-1]) + ",1\n")
```
